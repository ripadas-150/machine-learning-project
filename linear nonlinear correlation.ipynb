```python

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

generating linear data

```python
x_linear = np.linspace(0,10,50)
y_linear = 2*x_linear+5+np.random.rand(50)*2
```

```python
df=pd.DataFrame(x_linear)
df.head(10)
```

```python
df1=pd.DataFrame(y_linear)
df1.head(10)
```

generating nonlinear data

```python
x_nonlinear = np.linspace(0,10,50)
y_nonlinear = 2*np.sin(x_nonlinear)+ np.random.rand(50)
```

```python
df=pd.DataFrame(x_nonlinear)
df.head(10)
```

```python
df=pd.DataFrame(y_nonlinear)
df.head(10)
```

linear

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(x_linear,y_linear,color="blue",label="Data with linear relationship")
plt.plot(x_linear,2*x_linear+5,color="red",label="True linear relationship")
plt.xlabel("x")
plt.ylabel("y")
plt.title("nonLinear Relationship")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(x_nonlinear,y_nonlinear,color="green",label="Data with nonlinear relationship")
plt.plot(x_nonlinear,2*np.sin(x_nonlinear),color="black",label="True nonlinear relationship")
plt.xlabel("x")
plt.ylabel("y")
plt.title("nonLinear Relationship")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

correlation coeFFicient

```python
correlation_linear = np.corrcoef(x_linear,y_linear)[0,1]
correlation_nonlinear = np.corrcoef(x_nonlinear,y_nonlinear)[0,1]
```

linear regression

```python
reg1 = LinearRegression()
reg1
```

```python
reg1.fit(x_linear.reshape(-1,1),y_linear)
```

```python
reg1.score(x_linear.reshape(-1,1),y_linear)
```

```python
plt.plot(x_linear,reg1.predict(x_linear.reshape(-1,1)))
```

```python
plt.plot(x_linear,reg1.predict(x_linear.reshape(-1,1)),color="green",label="Best Fit line")
plt.scatter(x_linear,y_linear,color="blue",label="Data with linear relationship")

plt.xlabel("x")
plt.ylabel("y")
plt.title("nonLinear Relationship")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```
