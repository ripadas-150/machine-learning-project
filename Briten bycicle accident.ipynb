# one hot encoding

```python
Road_conditions = pd.get_dummies(x["Road_conditions"],dtype=int)
Road_conditions.head(10)
```

```python
Weather_conditions = pd.get_dummies(x["Weather_conditions"],dtype=int)
Weather_conditions.head(10)
```

```python
Road_type = pd.get_dummies(x["Road_type"],prefix="encoded",dtype=int)
Road_type.head(10)
```

```python
Light_conditions = pd.get_dummies(x["Light_conditions"],prefix="encoded",dtype=int)
Light_conditions.head(10)
```

# drop encoded properties

```python
x = df.drop(["Road_conditions","Light_conditions","Road_type","Weather_conditions","index","Date","Day","Accident_Index","Time"],axis = 1)
x.head(10)
```

# concate properties

```python
x=pd.concat([x,Road_conditions,Light_conditions,Road_type,Weather_conditions],axis=1)
x.head(10)
```

# train_test_split

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(xtrain,ytrain)
```

# predict the test result

```python
xtest
```

```python
ytest
```

# prediction

```python
pred = regressor.predict(xtest)
pred
```

# R squared value

```python
regressor.score(xtest,ytest)
```

```python
from sklearn.metrics import r2_score
score = r2_score(ytest,pred)
score
```
